<div>
    <h3>1.2. Outlier Detection: Statistics </h3>
</div>

<div>
    <h4><li>Using the Interquartile Range (IQR):</li></h4>
    <p>Outliers are typically defined as values that fall outside the range:</p>
    <pre>
        Lower Bound = Q1 - 1.5 * IQR
        Upper Bound = Q3 + 1.5 * IQR
    </pre> </div>
    <div>
    <h4><li>Using Standard Deviation (Z-Score Method):</li></h4>
    <p>A data point is considered an outlier if the absolute value of the Z-score is greater than a threshold (commonly 2 or 3).</p>
    <pre>
       |(x-μ)/σ| > 2 or 3
    </pre>
    <h4><li>Modified Z-Score (for smaller datasets):</li></h4>
    <p>For smaller datasets or when the data is not normally distributed, the modified Z-score can be used:</p>
    <pre>
        M = 0.6745 * (X - Median) / MAD
    </pre>
    <p>Where:</p>
    <ul>
        <li><strong>X</strong> is the data point.</li>
        <li><strong>Median</strong> is the median of the dataset.</li>
        <li><strong>MAD</strong> is the Median Absolute Deviation.</li>
    </ul>
    <p>A value of <strong>M > 3.5</strong> is typically considered an outlier.</p>
</div>

```python
# How to calculate the number of outliers for each column

def outliers_IQR(column_name):
    """Find outliers using IQR method"""
    
    # Check if the column is numeric
    if column_name.dtype not in ["int64", "float64"]:
        return  # Skip non-numeric columns
    
    # Compute Q1, Q3, and IQR
    q1 = column_name.quantile(0.25)
    q3 = column_name.quantile(0.75)
    iqr = q3 - q1
    
    # Define lower and upper bounds
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    
    # Find outliers 
    out = column_name[(column_name > upper) | (column_name < lower)]
    
    # Print outliers
    print(f"Column: {column_name.name} - Outliers: {list(out)}")  

# Apply to all numeric columns
for col in heart.columns:
    outliers_IQR(heart[col])
```

```python
outliers = []  # Store outliers

def z_out(hc):
    # Check if column is numeric
    if hc.dtype not in ["int64", "float64"]:
        return
    
    mean = hc.mean()  # Compute mean
    std_dev = hc.std()  # Compute standard deviation
    
    for value in hc:
        z_score = (value - mean) / std_dev  # Calculate Z-score
        if abs(z_score) > 3:  # Check for outliers
            outliers.append(value)  # Append actual outlier values
    
    print("Outliers are:", outliers)

# Call the function for a specific column
z_out(heart["Age (age in year)"])
```

```python
outliers_dict = {}  # Dictionary to store outliers for each column

def z_out(hc, column_name):
    # Check if column is numeric
    if hc.dtype not in ["int64", "float64"]:
        return
    
    mean = hc.mean()  # Compute mean
    std_dev = hc.std()  # Compute standard deviation
    
    column_outliers = []  # Store outliers for this column
    
    for value in hc:
        z_score = (value - mean) / std_dev  # Calculate Z-score
        if abs(z_score) > 3:  # Check for outliers
            column_outliers.append(value)  # Append actual outlier values
    
    # Store in dictionary if there are any outliers
    if column_outliers:
        outliers_dict[column_name] = column_outliers

# Iterate over all numeric columns in the DataFrame
for j in heart.columns:
    z_out(heart[j], j)

# Print the outliers for each column separately
for col, outlier_values in outliers_dict.items():
    print(f"Outliers in column '{col}': {outlier_values}")
```

```python
import pandas as pd

def find_outliers_zscore(series):
    """Returns a list of outliers in a given numeric pandas Series using the Z-score method."""
    if series.dtype not in ["int64", "float64"]:
        return []  # Skip non-numeric columns
    
    mean, std_dev = series.mean(), series.std()
    z_scores = (series - mean) / std_dev  # Compute Z-scores
    
    return series[abs(z_scores) > 3].tolist()  # Return outliers as a list

# Apply function to each numeric column and store results in a dictionary
outliers_dict = heart.select_dtypes(include=["int64", "float64"]).apply(find_outliers_zscore).to_dict()

# Print only columns that have outliers
outliers_dict = {col: vals for col, vals in outliers_dict.items() if vals}  # Remove empty lists

# Display results
for col, outliers in outliers_dict.items():
    print(f"Outliers in '{col}': {outliers}")
```
