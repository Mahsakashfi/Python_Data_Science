<p><h3>Adjusting Data Types</h3>

In our dataset, all features are currently stored as either `int64` or `float64`.  
However, not all of them are truly continuous numerical variables.  

- **Numerical variables** → mean, median, std, correlations, histograms, boxplots.  
  Examples: `Age`, `blood pressure`, `cholestoral`, `heart rate`, `depression`.  

- **Categorical variables** → counts, proportions, cross-tabs, bar plots.  
  Examples: `sex`, `chest pain`, `blood sugar`, `electrocardiographic`, `exercise induced`, 
  `slope`, `ca`, `thal`, `c` (target).  

To ensure correct analysis, we convert categorical variables to `category` dtype.
</p>

<h3>Outliers Detection</h3>
<p>1. Statistics - IQR</p>

```python
# Initialize an empty set to store all outlier row indices
all_outlier_indices = set()

# Loop over numeric columns
for col in data.select_dtypes(include='number').columns:
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    IQR = q3 - q1
    
    # Find outliers in this column
    col_outliers = data[(data[col] < q1 - 1.5 * IQR) | (data[col] > q3 + 1.5 * IQR)].index
    
    # Add these row indices to the set
    all_outlier_indices.update(col_outliers)

# Convert set to list
all_outlier_indices = list(all_outlier_indices)

# Create a DataFrame with all outlier rows
outliers_df = data.loc[all_outlier_indices]

outliers_df
```

<p>2. Statistics - Z-Score</p>

```python
lst = []
for i in data.select_dtypes(include='number').columns:
    outliers = data[((data[i] - data[i].mean())/data[i].std()).abs()>3]
    if not outliers.empty:
        lst.append({'column name': i, 
                 'outliers': outliers})
lst
```

<h4>Since the data is skewed for blood pressure','cholestoral', 'heart rate','depression', IQR is more robust than Z-score.</h4>

### Feature Scaling & Validation Strategy

We apply two approaches:

1. **Train-Test Split**  
   - Drop the target column `c`.  
   - Split the dataset into training (80%) and test (20%).  
   - Normalize the training set (fit the scaler) and apply the same transformation to the test set.  
   - Train ML models on the scaled training data and evaluate on the test data.

2. **Cross-Validation**  
   - Drop the target column `c`.  
   - Use cross-validation for more reliable performance estimation.  
   - Scaling is applied inside each fold (using `Pipeline`) so that the scaler is fitted only on training folds and applied to validation folds.  
   - This avoids data leakage and ensures fair evaluation.


```python
from sklearn.model_selection import train_test_split

X = data_encoded.drop('c', axis=1)   # features
y = data_encoded['c']                # target

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2020)
#stratify=y
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Initialize scaler
sc = MinMaxScaler()

# Fit on train only
X_train_scaled = pd.DataFrame(sc.fit_transform(X_train), columns=X_train.columns, index=X_train.index)

# Transform test with the same scaler
X_test_scaled = pd.DataFrame(sc.transform(X_test), columns=X_test.columns, index=X_test.index)
```
